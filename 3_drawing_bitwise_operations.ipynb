```python
import imageio.v3 as iio
import ipympl
import matplotlib.pyplot as plt
import numpy as np
import skimage as ski

%matplotlib widget
```

```python
maize_seedlings = iio.imread(uri="tutorial_figures/maize-seedlings.tif")

fig, ax = plt.subplots()

mask = np.ones(shape=maize_seedlings.shape[0:2], dtype="bool")
rr, cc = ski.draw.rectangle(start=(357, 44), end=(740, 720))
mask[rr, cc] = False

maize_seedlings[mask] = 0

ax.imshow(maize_seedlings)

fig, ax = plt.subplots()
ax.imshow(mask, cmap = "gray")

```

```python
canvas = np.zeros(shape=(600, 800, 3), dtype="uint8")

fig, ax = plt.subplots()


rr, cc = ski.draw.disk(center=(100, 100), radius=50)
canvas[rr, cc] = (255, 255, 255)
rr, cc = ski.draw.rectangle(start=(50,50), end=(110, 200))
canvas[rr, cc] = (255, 125, 0)
rr, cc = ski.draw.line(r0=100, c0=500, r1= 300, c1=100)
canvas[rr, cc] = (175, 178, 255)


ax.imshow(canvas)
```

```python
daylily = iio.imread(uri="tutorial_figures/daylily.jpg")

fig, ax = plt.subplots()

mask = np.zeros(shape=daylily.shape[0:2], dtype="bool")
rr, cc = ski.draw.rectangle(start=(320, 620), end=(460, 760))

mask[rr, cc] = True
daylily[mask] = (250, 125, 0)

ax.imshow(daylily)
```

```python
wellplates = iio.imread(uri="tutorial_figures/wellplate-01.jpg")
wellplates = np.array(wellplates)

mask = np.ones(shape=wellplates.shape[0:2], dtype="bool")

with open('tutorial_figures/centers.txt', 'r') as file:
    for line in file:
        center = line.split()
        center = [int(center[0]), int(center[1])]
        rr, cc = ski.draw.disk(center=(center[1], center[0]), radius = 20)
        mask[rr, cc] = False

wellplates[mask] = 0

fig, ax = plt.subplots()

ax.imshow(wellplates)
        
```

```python
wellplates = iio.imread(uri="tutorial_figures/wellplate-01.jpg")
wellplates = np.array(wellplates)

mask = np.ones(shape=wellplates.shape[0:2], dtype="bool")


cy = 108
for i in range(12):
    cx = 91
    for j in range(8):
        rr, cc = ski.draw.disk(center=[cy, cx], radius=20)
        cx = cx + 70
        mask[rr, cc] = False
    cy = cy + 72
    

wellplates[mask] = 0

fig, ax = plt.subplots()

ax.imshow(wellplates)
        
```
