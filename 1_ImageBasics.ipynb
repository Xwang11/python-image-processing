```python
import imageio.v3 as iio
import ipympl
import matplotlib.pyplot as plt
import numpy as np
import skimage as ski
```

```python
eight = iio.imread(uri="tutorial_figures/eight.tif")
print(type(eight))
```

```python
fig, ax = plt.subplots()

# Changing to 0
eight[2,1] = 1

print(eight.shape)
print(eight)

ax.imshow(eight)
```

```python
fig, ax = plt.subplots()
five = iio.imread(uri="tutorial_figures/eight.tif")
five[1,2] = 1
five[3,0] = 1

ax.imshow(five)
```

```python
three_colors = iio.imread(uri="tutorial_figures/eight.tif")

three_colors = three_colors * 128

three_colors[2,:] = 225.

fig, ax = plt.subplots()
ax.imshow(three_colors)
ax.imshow(three_colors, cmap=plt.cm.gray)
print(three_colors)
```

```python
randomizer = np.random.RandomState(2021)
checkerboard = randomizer.randint(0,255, size = (4,4,3))

fig, ax = plt.subplots()
ax.imshow(checkerboard)
print(checkerboard)
```

```python
upper_right_square = checkerboard[1,3,:]
print(upper_right_square)
```

```python
red_channel = checkerboard * [1, 0, 0]
fig, ax = plt.subplots()
ax.imshow(red_channel)

green_channel = checkerboard * [0, 1, 0]
fig, ax = plt.subplots()
ax.imshow(green_channel)

blue_channel = checkerboard * [0, 0, 1]
fig, ax = plt.subplots()
ax.imshow(blue_channel)
```

```python
# 1. Pure red
# 2. Pure green
# 3. Pure Blue
# 4. White
# 5. Black
# 6. Grey

# A 5000x5000 = 25000000 pixels
# Each pixel is 24 bits b/c using 24bit RGB color model
# Size = 25000000 * 24 = 600000000 bits

```

```python
dim = 5000

img = np.zeros((dim, dim, 3), dtype = "uint8")
img.fill(255)
iio.imwrite(uri="ws.bmp", image=img)
iio.imwrite(uri="ws.jpeg", image=img)
```
