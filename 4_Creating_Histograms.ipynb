```python
import imageio.v3 as iio
import ipympl
import matplotlib.pyplot as plt
import numpy as np
import skimage as ski

%matplotlib widget
```

```python
plant_seedling = iio.imread(uri="tutorial_figures/plant-seedling.jpg", mode="L")

plant_seedling = ski.util.img_as_float(plant_seedling)

histogram, bin_edges = np.histogram(plant_seedling, bins=256, range=(0,1))

fig, ax = plt.subplots()
ax.imshow(plant_seedling, cmap="gray")

fig, ax = plt.subplots()
ax.set_title("Grayscale Histogram")
ax.set_xlabel("Grayscale Value")
ax.set_ylabel("Pixel Count")
ax.set_xlim([0.0, 1.0])

ax.plot(bin_edges[0:-1], histogram)
```

```python
plant_seedling = iio.imread(uri="tutorial_figures/plant-seedling.jpg", mode="L")
plant_seedling = ski.util.img_as_float(plant_seedling)

mask = np.zeros(shape=plant_seedling.shape[0:2], dtype="bool")
rr, cc = ski.draw.rectangle(start=(200, 425), end=(380, 480))
mask[rr, cc] = True

fig, ax = plt.subplots()
ax.imshow(mask, cmap="gray")

histogram, bin_edges = np.histogram(plant_seedling[mask], bins=256, range=(0,1))
fig, ax = plt.subplots()
ax.set_title("Grayscale Histogram")
ax.set_xlabel("Grayscale Value")
ax.set_ylabel("Pixel Count")
ax.set_xlim([0.0, 1.0])

ax.plot(bin_edges[0:-1], histogram)
```

```python
plant_seedling = iio.imread(uri="tutorial_figures/plant-seedling.jpg")

fig, ax = plt.subplots()
ax.imshow(plant_seedling)

mask=np.zeros(shape = plant_seedling.shape[0:2], dtype="bool")
rr, cc = ski.draw.rectangle(start=(199, 410), end=(384, 484))
mask[rr, cc] = True

colors = ("red", "green", "blue")
fig, ax = plt.subplots()
ax.set_xlim([0, 256])

for channel_id, color in enumerate(colors):
    histogram, bin_edges = np.histogram(
        plant_seedling[:, :, channel_id], bins=256, range=(0, 256)
    )
    ax.plot(bin_edges[0:-1], histogram, color=color)

ax.set_title("Color Histogram")
ax.set_xlabel("Color Value")
ax.set_ylabel("Pixel Count")
```

```python
wellplate = iio.imread(uri="tutorial_figures/wellplate-02.tif")

fig, ax = plt.subplots()

ax.imshow(wellplate)

radius = 50
center = [240, 1050]

mask = np.zeros(shape=wellplate.shape[0:2], dtype="bool")
rr, cc = ski.draw.disk(center=center, radius = radius)
mask[rr, cc] = True

wellplate[~mask] = 0

fig, ax = plt.subplots()
ax.imshow(wellplate)


# Color Histograph

colors = ("red", "green", "blue")
fig, ax = plt.subplots()
ax.set_xlim([0, 256])

for channel_id, color in enumerate(colors):
    histogram, bin_edges = np.histogram(
        wellplate[mask, channel_id], bins=256, range=(0, 256)
    )
    ax.plot(bin_edges[0:-1], histogram, color=color)

ax.set_title("Color Histogram")
ax.set_xlabel("Color Value")
ax.set_ylabel("Pixel Count")
```
