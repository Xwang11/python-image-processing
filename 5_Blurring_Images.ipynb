```python
import imageio.v3 as iio
import ipympl
import matplotlib.pyplot as plt
import skimage as ski

%matplotlib widget
```

```python
# Gaussian Blur, ie, Low-Pass blurring filter that smooths edges and remove noise from an image. 

image = iio.imread(uri="tutorial_figures/gaussian-original.png")

#display image
fig, ax = plt.subplots()
ax.imshow(image)

# Standered deviation
sigma = 3.0

# Kernal Size = 2 * sigma * truncate

print(image.ndim)

#Apply Gaussian Blur, creating a new image
blurred = ski.filters.gaussian(
    image, sigma=(sigma, sigma), truncate=3.5, channel_axis=-1
)

fig, ax = plt.subplots()
ax.imshow(blurred)
```

```python
# read image in gray scale

image = iio.imread('tutorial_figures/colonies-01.tif')
image_gray = ski.color.rgb2gray(image)

# Define the pixels we want to view the intensity -> one line
xmin, xmax = (0, image_gray.shape[1])
Y = ymin = ymax = 150

# View image
fig, ax = plt.subplots()
ax.imshow(image_gray, cmap='gray')
ax.plot([xmin, xmax], [ymin, ymax], color="red")
```

```python
image_gray_pixels_slice = image_gray[Y, :]

image_gray_pixels_slice = ski.img_as_ubyte(image_gray_pixels_slice)

fig, ax = plt.subplots()
ax.plot(image_gray_pixels_slice, color='red')
ax.set_ylim(255, 0)
ax.set_ylabel('L')
ax.set_xlabel('X')
```

```python
image_blur = ski.filters.gaussian(image_gray, sigma=3)

image_blur_pixels_slice = image_blur[Y, :]
image_blur_pixels_slice = ski.img_as_ubyte(image_blur_pixels_slice)

fig, ax = plt.subplots()
ax.plot(image_blur_pixels_slice, "red")
ax.set_ylim(255, 0)
ax.set_ylabel('L')
ax.set_xlabel('X')
```
